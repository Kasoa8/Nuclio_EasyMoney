```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
```

```python
# Ruta al archivo Parquet 
ruta_archivo = "/Users/fernando/Desktop/TFM/Nuclio_EasyMoney/0_Data/clustering_3_result.csv"
# Leer el archivo Parquet 
df_cluster = pd.read_csv(ruta_archivo)
```

```python
df_cluster
```

```python
# Suponiendo que queremos predecir la compra de un producto 'credit_card'
df_cluster['purchased_credit_card'] = df_cluster['credit_card'].apply(lambda x: 1 if x > 0 else 0)
```

```python
features = ["salary", "loan_amount", "debt_to_income_ratio", "account_balance", 
"number_of_products", "product_engagement_score", "payroll_account", "emc_account", "em_account_p", "em_acount", "credit_card",
"age", "segment", "active_customer", "region_code"
]
X = df_cluster[features]
y = df_cluster['purchased_credit_card']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]

print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
```

```python
importances = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

print("Feature Importances:\n", feature_importance)
```
